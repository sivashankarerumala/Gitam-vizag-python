```python
import pandas as pd      ## convertion dict to series
internal1= {'s1':35,'s2':34,'s3':33}
internal1= pd.Series(internal1)
internal2={'s1':23,'s2':34,'s3':35}
internal2= pd.Series(internal2)
print(internal1)
print(internal2)
print(type(internal2))
```

```python
final = {'internal1':internal1,'internal2':internal2}
final = pd.DataFrame(final)
print(final)
print(type(final))
```
