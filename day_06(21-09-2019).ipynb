### Visualization Data
- An interesting way to display Dataframes
- With matplotlib

```python
import  matplotlib.pyplot as plt
a1=[1,2,3,4]
a2=[1,2,3,4]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import  matplotlib.pyplot as plt
a1=[1,2,3,4]
a2=[5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='green',marker='*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```
